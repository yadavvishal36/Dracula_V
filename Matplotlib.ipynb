```python
import numpy as np 
from matplotlib import pyplot as plt 


start = 1
end = 1*60*60 #3600
noofpoints = 3600

X = np.linspace(start,end,noofpoints) # x axis on time 
pi = 22/7
omega = int(input("enter frequency"))

Y = np.sin(2 * pi * omega * X)
plt.plot(X,Y)
plt.xlabel("time ($s$)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
Y2 = np.sin(Y)
```
