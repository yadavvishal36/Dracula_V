1 . Implement the Viterbi algorithm in python to find the most likely sequence of hidden states given an observed sequence and an HMM . 


```python
import numpy as np

def viterbi(observations, states, start_prob, trans_prob, emit_prob):
    # Number of states
    num_states = len(states)
    # Number of observations
    num_obs = len(observations)

    # Viterbi matrix where rows are states and columns are observations
    viterbi_matrix = np.zeros((num_states, num_obs))
    
    # Matrix to store the optimal path (backtrack)
    backtrack_matrix = np.zeros((num_states, num_obs), dtype=int)

    # Initialization step
    for s in range(num_states):
        viterbi_matrix[s, 0] = start_prob[s] * emit_prob[s][observations[0]]
        backtrack_matrix[s, 0] = 0

    # Recursion step
    for t in range(1, num_obs):
        for s in range(num_states):
            transition_probs = [viterbi_matrix[prev_s, t-1] * trans_prob[prev_s][s] for prev_s in range(num_states)]
            max_prob = max(transition_probs)
            max_state = np.argmax(transition_probs)
            viterbi_matrix[s, t] = max_prob * emit_prob[s][observations[t]]
            backtrack_matrix[s, t] = max_state

    # Termination step
    best_last_state = np.argmax(viterbi_matrix[:, -1])
    best_path_prob = viterbi_matrix[best_last_state, -1]

    # Backtracking to find the best path
    best_path = [best_last_state]
    for t in range(num_obs - 1, 0, -1):
        best_last_state = backtrack_matrix[best_last_state, t]
        best_path.insert(0, best_last_state)

    return best_path, best_path_prob

# Example usage

# States (hidden states)
states = ['Healthy', 'Fever']

# Observations (encoded as integers)
observations = [0, 1, 1]  # Example: 'Normal', 'Cold', 'Cold'

# Observation types (for example, Normal=0, Cold=1, Dizzy=2)
obs_map = {0: 'Normal', 1: 'Cold', 2: 'Dizzy'}

# Start probability (probability of starting in each state)
start_prob = [0.6, 0.4]

# Transition probability (probability of moving from one state to another)
trans_prob = [
    [0.7, 0.3],  # From Healthy to (Healthy, Fever)
    [0.4, 0.6]   # From Fever to (Healthy, Fever)
]

# Emission probability (probability of observing each observation given a state)
emit_prob = [
    [0.5, 0.4, 0.1],  # Healthy emits (Normal, Cold, Dizzy)
    [0.1, 0.3, 0.6]   # Fever emits (Normal, Cold, Dizzy)
]

# Run the Viterbi algorithm
best_path, best_path_prob = viterbi(observations, states, start_prob, trans_prob, emit_prob)

# Output the result
best_path_states = [states[state] for state in best_path]
print("Most likely sequence of states:", best_path_states)
print("Probability of the best path:", best_path_prob)
```

2 . write a python program to simulate and HMM with given transition and emission matrices and generate a sequence of observations 

```python
import numpy as np

def simulate_hmm(states, observations, start_prob, trans_prob, emit_prob, num_steps):
    # Convert the number of states and observations to more readable form
    num_states = len(states)
    num_observations = len(observations)
    
    # List to store the generated states and observations
    generated_states = []
    generated_observations = []
    
    # Choose the initial state based on the start probabilities
    current_state = np.random.choice(range(num_states), p=start_prob)
    generated_states.append(current_state)
    
    # Generate the first observation based on the initial state
    current_observation = np.random.choice(range(num_observations), p=emit_prob[current_state])
    generated_observations.append(current_observation)
    
    # Generate subsequent states and observations
    for _ in range(1, num_steps):
        # Transition to the next state based on the transition probabilities of the current state
        next_state = np.random.choice(range(num_states), p=trans_prob[current_state])
        generated_states.append(next_state)
        
        # Generate the next observation based on the new state
        next_observation = np.random.choice(range(num_observations), p=emit_prob[next_state])
        generated_observations.append(next_observation)
        
        # Move to the next state
        current_state = next_state
    
    # Convert states and observations to their respective labels
    generated_state_labels = [states[s] for s in generated_states]
    generated_observation_labels = [observations[o] for o in generated_observations]
    
    return generated_state_labels, generated_observation_labels

# Example usage:

# Hidden states
states = ['Rainy', 'Sunny']

# Observations
observations = ['Walk', 'Shop', 'Clean']

# Start probabilities (initial distribution over the hidden states)
start_prob = [0.6, 0.4]  # 60% chance of starting in 'Rainy', 40% in 'Sunny'

# Transition probability matrix
trans_prob = [
    [0.7, 0.3],  # Transition from 'Rainy' to ('Rainy', 'Sunny')
    [0.4, 0.6]   # Transition from 'Sunny' to ('Rainy', 'Sunny')
]

# Emission probability matrix (observation likelihood given the state)
emit_prob = [
    [0.1, 0.4, 0.5],  # 'Rainy' emits (Walk: 10%, Shop: 40%, Clean: 50%)
    [0.6, 0.3, 0.1]   # 'Sunny' emits (Walk: 60%, Shop: 30%, Clean: 10%)
]

# Number of steps to simulate
num_steps = 10

# Simulate the HMM and generate a sequence of states and observations
generated_states, generated_observations = simulate_hmm(states, observations, start_prob, trans_prob, emit_prob, num_steps)

# Output the result
print("Generated Hidden States:", generated_states)
print("Generated Observations:", generated_observations)
```

```python
import numpy as np

def forward_algorithm(observations, states, start_prob, trans_prob, emit_prob):
    """
    Forward algorithm to calculate the probability of an observed sequence.
    
    Parameters:
    - observations: A list of observations (encoded as integers).
    - states: A list of hidden states.
    - start_prob: Initial probabilities for the states.
    - trans_prob: Transition probability matrix.
    - emit_prob: Emission probability matrix.
    
    Returns:
    - forward_prob: The probability of the observation sequence given the HMM.
    """
    # Number of states
    num_states = len(states)
    
    # Number of observations
    num_obs = len(observations)
    
    # Initialize the forward matrix
    forward_matrix = np.zeros((num_states, num_obs))
    
    # Initialization step (t = 0)
    for s in range(num_states):
        forward_matrix[s, 0] = start_prob[s] * emit_prob[s][observations[0]]
    
    # Recursion step (t > 0)
    for t in range(1, num_obs):
        for s in range(num_states):
            forward_matrix[s, t] = sum(forward_matrix[prev_s, t-1] * trans_prob[prev_s][s] for prev_s in range(num_states)) * emit_prob[s][observations[t]]
    
    # Termination step: Sum the probabilities of all states at the final time step
    forward_prob = sum(forward_matrix[s, num_obs - 1] for s in range(num_states))
    
    return forward_prob

# Example usage:

# Hidden states
states = ['Rainy', 'Sunny']

# Observations (encoded as integers: 0=Walk, 1=Shop, 2=Clean)
observations = [0, 1, 2]  # Sequence of observations ('Walk', 'Shop', 'Clean')

# Start probabilities
start_prob = [0.6, 0.4]  # 60% chance of starting in 'Rainy', 40% in 'Sunny'

# Transition probability matrix
trans_prob = [
    [0.7, 0.3],  # From 'Rainy' to ('Rainy', 'Sunny')
    [0.4, 0.6]   # From 'Sunny' to ('Rainy', 'Sunny')
]

# Emission probability matrix (probability of observing each observation given a state)
emit_prob = [
    [0.1, 0.4, 0.5],  # 'Rainy' emits ('Walk': 10%, 'Shop': 40%, 'Clean': 50%)
    [0.6, 0.3, 0.1]   # 'Sunny' emits ('Walk': 60%, 'Shop': 30%, 'Clean': 10%)
]

# Calculate the probability of the observation sequence using the forward algorithm
forward_prob = forward_algorithm(observations, states, start_prob, trans_prob, emit_prob)

# Output the result
print("Probability of the observed sequence:", forward_prob)
```

4 . Write a python script to perform a Baum-Welch algorithm for training an HMM on a given set of observed sequence 

```python
import numpy as np 

def forward_algorithm(observations, states, start_prob, trans_prob, emit_prob):
    """
    Forward algorithm to calculate the forward probabilities.
    Returns the forward probabilities and the total probability of the sequence.
    """
    num_states = len(states)
    num_obs = len(observations)
    
    # Forward matrix
    forward_matrix = np.zeros((num_states, num_obs))
    
    # Initialization
    for s in range(num_states):
        forward_matrix[s, 0] = start_prob[s] * emit_prob[s, observations[0]]
    
    # Recursion
    for t in range(1, num_obs):
        for s in range(num_states):
            forward_matrix[s, t] = sum(forward_matrix[prev_s, t-1] * trans_prob[prev_s, s] for prev_s in range(num_states)) * emit_prob[s, observations[t]]
    
    # Total probability
    total_prob = sum(forward_matrix[s, num_obs - 1] for s in range(num_states))
    
    return forward_matrix, total_prob

def backward_algorithm(observations, states, trans_prob, emit_prob):
    """
    Backward algorithm to calculate the backward probabilities.
    """
    num_states = len(states)
    num_obs = len(observations)
    
    # Backward matrix
    backward_matrix = np.zeros((num_states, num_obs))
    
    # Initialization
    for s in range(num_states):
        backward_matrix[s, num_obs - 1] = 1  # Last time step
    
    # Recursion
    for t in range(num_obs - 2, -1, -1):
        for s in range(num_states):
            backward_matrix[s, t] = sum(trans_prob[s, next_s] * emit_prob[next_s, observations[t+1]] * backward_matrix[next_s, t+1] for next_s in range(num_states))
    
    return backward_matrix

def baum_welch(observations, states, start_prob, trans_prob, emit_prob, num_iterations=100):
    """
    Baum-Welch algorithm to train an HMM.
    """
    num_states = len(states)
    num_obs = len(observations)
    
    for n in range(num_iterations):
        # Expectation Step: Calculate forward and backward probabilities
        forward_matrix, total_prob = forward_algorithm(observations, states, start_prob, trans_prob, emit_prob)
        backward_matrix = backward_algorithm(observations, states, trans_prob, emit_prob)
        
        # Gamma and Xi (expected counts)
        gamma = np.zeros((num_states, num_obs))
        xi = np.zeros((num_states, num_states, num_obs - 1))
        
        for t in range(num_obs):
            for s in range(num_states):
                gamma[s, t] = (forward_matrix[s, t] * backward_matrix[s, t]) / total_prob
        
        for t in range(num_obs - 1):
            denom = sum(forward_matrix[i, t] * trans_prob[i, j] * emit_prob[j, observations[t+1]] * backward_matrix[j, t+1]
                        for i in range(num_states) for j in range(num_states))
            for i in range(num_states):
                for j in range(num_states):
                    xi[i, j, t] = (forward_matrix[i, t] * trans_prob[i, j] * emit_prob[j, observations[t+1]] * backward_matrix[j, t+1]) / denom
        
        # Maximization Step: Re-estimate the model parameters
        # Re-estimate initial probabilities
        for s in range(num_states):
            start_prob[s] = gamma[s, 0]
        
        # Re-estimate transition probabilities
        for i in range(num_states):
            for j in range(num_states):
                trans_prob[i, j] = sum(xi[i, j, t] for t in range(num_obs - 1)) / sum(gamma[i, t] for t in range(num_obs - 1))
        
        # Re-estimate emission probabilities
        for s in range(num_states):
            for k in range(emit_prob.shape[1]):
                emit_prob[s, k] = sum(gamma[s, t] for t in range(num_obs) if observations[t] == k) / sum(gamma[s, t] for t in range(num_obs))
    
    return start_prob, trans_prob, emit_prob

# Example usage:

# Hidden states
states = ['Rainy', 'Sunny']

# Observations (encoded as integers: 0=Walk, 1=Shop, 2=Clean)
observations = [0, 1, 2, 1, 0, 2, 0]  # Sequence of observations ('Walk', 'Shop', 'Clean')

# Start probabilities (initial guess)
start_prob = np.array([0.6, 0.4])

# Transition probability matrix (initial guess)
trans_prob = np.array([
    [0.7, 0.3],  # Transition from 'Rainy' to ('Rainy', 'Sunny')
    [0.4, 0.6]   # Transition from 'Sunny' to ('Rainy', 'Sunny')
])

# Emission probability matrix (initial guess)
emit_prob = np.array([
    [0.1, 0.4, 0.5],  # 'Rainy' emits (Walk: 10%, Shop: 40%, Clean: 50%)
    [0.6, 0.3, 0.1]   # 'Sunny' emits (Walk: 60%, Shop: 30%, Clean: 10%)
])

# Train the HMM using the Baum-Welch algorithm
num_iterations = 100
new_start_prob, new_trans_prob, new_emit_prob = baum_welch(observations, states, start_prob, trans_prob, emit_prob, num_iterations)

# Output the re-estimated parameters
print("Re-estimated start probabilities:", new_start_prob)
print("Re-estimated transition probabilities:\n", new_trans_prob)
print("Re-estimated emission probabilities:\n", new_emit_prob)
```

6. Develop a python program to estimate the parameters ( transition and emission matrices ) of an HMM using the expectation and maximization algorithm  

```python
import numpy as np

# Forward step
def forward(obs_seq, A, B, pi):
    N = len(A)  # Number of states
    T = len(obs_seq)  # Length of observation sequence
    alpha = np.zeros((T, N))
    
    # Initialization
    alpha[0, :] = pi * B[:, obs_seq[0]]
    
    # Recursion
    for t in range(1, T):
        for j in range(N):
            alpha[t, j] = np.sum(alpha[t-1, :] * A[:, j]) * B[j, obs_seq[t]]
    
    return alpha

# Backward step
def backward(obs_seq, A, B):
    N = len(A)  # Number of states
    T = len(obs_seq)  # Length of observation sequence
    beta = np.zeros((T, N))
    
    # Initialization
    beta[T-1, :] = 1  # Set beta at T-1 to 1
    
    # Recursion
    for t in reversed(range(T-1)):
        for i in range(N):
            beta[t, i] = np.sum(A[i, :] * B[:, obs_seq[t+1]] * beta[t+1, :])
    
    return beta

# Baum-Welch (EM) algorithm
def baum_welch(obs_seq, N, M, max_iters=100):
    T = len(obs_seq)
    
    # Randomly initialize A, B, and pi
    A = np.random.rand(N, N)
    A = A / A.sum(axis=1, keepdims=True)
    
    B = np.random.rand(N, M)
    B = B / B.sum(axis=1, keepdims=True)
    
    pi = np.random.rand(N)
    pi = pi / pi.sum()
    
    for iteration in range(max_iters):
        # E-step: Forward and Backward
        alpha = forward(obs_seq, A, B, pi)
        beta = backward(obs_seq, A, B)
        
        # Compute xi and gamma
        xi = np.zeros((T-1, N, N))
        gamma = np.zeros((T, N))
        
        for t in range(T-1):
            denom = np.sum(alpha[t, :] * np.dot(A, B[:, obs_seq[t+1]] * beta[t+1, :]))
            for i in range(N):
                gamma[t, i] = np.sum(alpha[t, i] * beta[t, i]) / denom
                for j in range(N):
                    xi[t, i, j] = (alpha[t, i] * A[i, j] * B[j, obs_seq[t+1]] * beta[t+1, j]) / denom
        
        gamma[-1, :] = alpha[-1, :] / np.sum(alpha[-1, :])  # Last gamma for t = T-1
        
        # M-step: Update parameters
        A = np.sum(xi, axis=0) / np.sum(gamma[:-1], axis=0, keepdims=True)
        B = np.copy(B)
        for j in range(N):
            for k in range(M):
                mask = (obs_seq == k)
                B[j, k] = np.sum(gamma[mask, j]) / np.sum(gamma[:, j])
        
        pi = gamma[0, :]
        
    return A, B, pi

# Example usage
if __name__ == "__main__":
    # Observation sequence (encoded as integers)
    obs_seq = np.array([0, 1, 0, 2, 1, 0])  # Example observation sequence
    N = 2  # Number of hidden states
    M = 3  # Number of observation symbols
    
    A, B, pi = baum_welch(obs_seq, N, M, max_iters=100)
    
    print("Estimated Transition Matrix (A):")
    print(A)
    print("\nEstimated Emission Matrix (B):")
    print(B)
    print("\nEstimated Initial Probabilities (pi):")
    print(pi)
```

7 .Write a python script that visualizes the state transitions of an HMM over time using a state transition diagram

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# Function to visualize the HMM state transitions
def visualize_hmm_transitions(A, states, show_labels = True):
    """
    Visualizes the state transition diagram of an HMM.

    Parameters:
    - A: Transition matrix (N x N) where N is the number of states
    - states: List of state names or identifiers
    - show_labels: Whether to show edge weights (transition probabilities) as labels
    """
    N = len(A)  # Number of states
    
    # Create a directed graph
    G = nx.DiGraph()
    
    # Add nodes (states)
    G.add_nodes_from(states)
    
    # Add edges with weights (transition probabilities)
    for i in range(N):
        for j in range(N):
            if A[i, j] > 0:  # Only add edges with non-zero probabilities
                G.add_edge(states[i], states[j], weight=A[i, j])
    
    # Generate layout for nodes
    pos = nx.spring_layout(G)
    
    # Draw nodes
    nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, font_size=12, font_weight='bold', arrows=True, arrowstyle='-|>', arrowsize=20)
    
    # Draw edges
    edges = G.edges(data=True)
    nx.draw_networkx_edges(G, pos, edgelist=edges, arrowstyle='-|>', arrowsize=20)
    
    # Optionally add edge labels (transition probabilities)
    if show_labels:
        edge_labels = {(u, v): f"{d['weight']:.2f}" for u, v, d in edges}
        nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10)
    
    # Display the graph
    plt.title("HMM State Transition Diagram")
    plt.show()

# Example usage
if __name__ == "__main__":
    # Transition matrix (example)
    A = np.array([[0.6, 0.4],
                  [0.5, 0.5]])  # 2 states with transition probabilities
    
    # List of states
    states = ['State 1', 'State 2']
    
    # Visualize the state transitions
    visualize_hmm_transitions(A, states, show_labels=True)
```

8. Implement a python function to compute the likelihood of a given observed sequence using both the forward and backward algorithms 

```python
import numpy as np

# Forward algorithm
def forward(obs_seq, A, B, pi):
    """
    Forward algorithm to compute the likelihood of an observation sequence.
    
    Parameters:
    - obs_seq: List or array of observed sequence (integers representing observations)
    - A: Transition matrix (N x N)
    - B: Emission matrix (N x M), where M is the number of observation symbols
    - pi: Initial state distribution (1D array of length N)
    
    Returns:
    - alpha: The forward probability matrix (T x N)
    - likelihood: The likelihood of the observation sequence
    """
    N = len(A)  # Number of states
    T = len(obs_seq)  # Length of observation sequence
    alpha = np.zeros((T, N))
    
    # Initialization step
    alpha[0, :] = pi * B[:, obs_seq[0]]
    
    # Recursion step
    for t in range(1, T):
        for j in range(N):
            alpha[t, j] = np.sum(alpha[t-1, :] * A[:, j]) * B[j, obs_seq[t]]
    
    # Termination step: Sum over all possible final states
    likelihood = np.sum(alpha[T-1, :])
    return alpha, likelihood

# Backward algorithm
def backward(obs_seq, A, B):
    """
    Backward algorithm to compute the likelihood of an observation sequence.
    
    Parameters:
    - obs_seq: List or array of observed sequence (integers representing observations)
    - A: Transition matrix (N x N)
    - B: Emission matrix (N x M), where M is the number of observation symbols
    
    Returns:
    - beta: The backward probability matrix (T x N)
    - likelihood: The likelihood of the observation sequence
    """
    N = len(A)  # Number of states
    T = len(obs_seq)  # Length of observation sequence
    beta = np.zeros((T, N))
    
    # Initialization step
    beta[T-1, :] = 1  # At the last time step, set all betas to 1
    
    # Recursion step
    for t in reversed(range(T-1)):
        for i in range(N):
            beta[t, i] = np.sum(A[i, :] * B[:, obs_seq[t+1]] * beta[t+1, :])
    
    # Termination step: Compute the likelihood
    likelihood = np.sum(pi * B[:, obs_seq[0]] * beta[0, :])
    return beta, likelihood

# Function to compute likelihood using both forward and backward algorithms
def compute_likelihood(obs_seq, A, B, pi):
    """
    Computes the likelihood of an observation sequence using both
    the forward and backward algorithms and verifies they give the same result.
    
    Parameters:
    - obs_seq: List or array of observed sequence (integers representing observations)
    - A: Transition matrix (N x N)
    - B: Emission matrix (N x M), where M is the number of observation symbols
    - pi: Initial state distribution (1D array of length N)
    
    Returns:
    - likelihood_forward: Likelihood of the observation sequence from the forward algorithm
    - likelihood_backward: Likelihood of the observation sequence from the backward algorithm
    """
    # Forward algorithm
    _, likelihood_forward = forward(obs_seq, A, B, pi)
    
    # Backward algorithm
    _, likelihood_backward = backward(obs_seq, A, B)
    
    return likelihood_forward, likelihood_backward

# Example usage
if __name__ == "__main__":
    # Transition matrix (example)
    A = np.array([[0.7, 0.3],
                  [0.4, 0.6]])  # 2 states with transition probabilities
    
    # Emission matrix (example)
    B = np.array([[0.5, 0.5],
                  [0.1, 0.9]])  # Emission probabilities for 2 states and 2 observations
    
    # Initial state distribution (example)
    pi = np.array([0.6, 0.4])  # Probability of starting in each state
    
    # Observation sequence (example)
    obs_seq = [0, 1, 0]  # Sequence of observations (0 and 1 represent different observation symbols)
    
    # Compute likelihood using both forward and backward algorithms
    likelihood_forward, likelihood_backward = compute_likelihood(obs_seq, A, B, pi)
    
    # Output the likelihood values
    print(f"Likelihood (Forward Algorithm): {likelihood_forward}")
    print(f"Likelihood (Backward Algorithm): {likelihood_backward}")
    
    # Both likelihoods should be the same (within numerical precision)
    assert np.isclose(likelihood_forward, likelihood_backward), "Likelihoods do not match!"
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate synthetic data from an HMM
def generate_hmm_data(A, B, pi, num_samples):
    """
    Generate synthetic data from an HMM.
    
    Parameters:
    - A: Transition matrix (N x N), where N is the number of states
    - B: Emission matrix (N x M), where M is the number of observation symbols
    - pi: Initial state distribution (1D array of length N)
    - num_samples: Number of samples (time steps) to generate
    
    Returns:
    - hidden_states: Sequence of hidden states
    - observations: Sequence of observations
    """
    N = len(A)  # Number of hidden states
    M = len(B[0])  # Number of possible observations
    
    # Initialize arrays to store the hidden states and observations
    hidden_states = np.zeros(num_samples, dtype=int)
    observations = np.zeros(num_samples, dtype=int)
    
    # Sample the initial state based on the initial state distribution (pi)
    hidden_states[0] = np.random.choice(N, p=pi)
    observations[0] = np.random.choice(M, p=B[hidden_states[0]])
    
    # Generate the hidden states and observations for each time step
    for t in range(1, num_samples):
        # Sample the next hidden state based on the transition matrix
        hidden_states[t] = np.random.choice(N, p=A[hidden_states[t-1]])
        
        # Sample the observation based on the emission matrix for the current hidden state
        observations[t] = np.random.choice(M, p=B[hidden_states[t]])
    
    return hidden_states, observations

# Function to visualize the hidden states and observations
def visualize_hmm_data(hidden_states, observations):
    """
    Visualize the hidden states and observations generated by an HMM.
    
    Parameters:
    - hidden_states: Sequence of hidden states
    - observations: Sequence of observations
    """
    num_samples = len(hidden_states)
    time_steps = np.arange(num_samples)
    
    # Create a figure with two subplots
    fig, ax = plt.subplots(2, 1, figsize=(10, 6), sharex=True)
    
    # Plot the hidden states
    ax[0].plot(time_steps, hidden_states, marker='o', linestyle='-', color='r')
    ax[0].set_title("Hidden States Over Time")
    ax[0].set_ylabel("Hidden State")
    ax[0].grid(True)
    
    # Plot the observations
    ax[1].plot(time_steps, observations, marker='o', linestyle='-', color='b')
    ax[1].set_title("Observations Over Time")
    ax[1].set_ylabel("Observation")
    ax[1].set_xlabel("Time Steps")
    ax[1].grid(True)
    
    # Display the plot
    plt.tight_layout()
    plt.show()

# Example usage
if __name__ == "__main__":
    # Define the HMM parameters
    
    # Transition matrix (example for 3 states)
    A = np.array([[0.7, 0.2, 0.1],
                  [0.3, 0.5, 0.2],
                  [0.1, 0.3, 0.6]])
    
    # Emission matrix (example for 3 states and 2 possible observations)
    B = np.array([[0.6, 0.4],
                  [0.3, 0.7],
                  [0.8, 0.2]])
    
    # Initial state distribution (example for 3 states)
    pi = np.array([0.5, 0.3, 0.2])
    
    # Number of samples to generate
    num_samples = 100
    
    # Generate synthetic data from the HMM
    hidden_states, observations = generate_hmm_data(A, B, pi, num_samples)
    
    # Visualize the generated data
    visualize_hmm_data(hidden_states, observations)
```

10. write a python function to compute the stationary distribution of an HMM using its transition matrix 

```python
#method 1 
import numpy as np

def compute_stationary_distribution(A):
    """
    Computes the stationary distribution of a Hidden Markov Model (HMM)
    using its transition matrix A.
    
    Parameters:
    - A: Transition matrix (N x N), where N is the number of states.
    
    Returns:
    - pi: Stationary distribution (1D array of length N).
    """
    # Ensure that A is a square matrix
    N = A.shape[0]
    
    # Subtract the identity matrix from the transition matrix (A - I)
    A_mod = A.T - np.eye(N)
    
    # Add a normalization constraint (sum of probabilities = 1)
    A_mod[-1, :] = 1  # Replace the last row with ones (normalization condition)
    
    # Create a b vector with the last element as 1 (for normalization)
    b = np.zeros(N)
    b[-1] = 1
    
    # Solve the linear system A_mod * pi = b
    pi = np.linalg.solve(A_mod, b)
    
    return pi

# Example usage
if __name__ == "__main__":
    # Transition matrix (example for 3 states)
    A = np.array([[0.7, 0.2, 0.1],
                  [0.3, 0.5, 0.2],
                  [0.2, 0.3, 0.5]])
    
    # Compute the stationary distribution
    pi = compute_stationary_distribution(A)
    
    # Output the stationary distribution
    print("Stationary Distribution:", pi)
```

```python
#method 2
def compute_stationary_distribution_power_method(A, tol=1e-8, max_iter=1000):
    """
    Computes the stationary distribution using the power method.
    
    Parameters:
    - A: Transition matrix (N x N)
    - tol: Convergence tolerance
    - max_iter: Maximum number of iterations
    
    Returns:
    - pi: Stationary distribution (1D array of length N)
    """
    N = A.shape[0]
    
    # Start with a uniform distribution
    pi = np.ones(N) / N
    
    # Iterate until convergence
    for _ in range(max_iter):
        pi_next = np.dot(pi, A)
        
        # Check for convergence
        if np.linalg.norm(pi_next - pi) < tol:
            break
        
        pi = pi_next
    
    return pi

# Example usage
if __name__ == "__main__":
    A = np.array([[0.7, 0.2, 0.1],
                  [0.3, 0.5, 0.2],
                  [0.2, 0.3, 0.5]])
    
    pi = compute_stationary_distribution_power_method(A)
    print("Stationary Distribution (Power Method):", pi)
```
